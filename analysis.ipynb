# **Data Analysis of Futbol Matches**

Neccessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
from utils.constants import PROFESSIONAL_ARGENTINA_LEAGUE
```

## **Brasileir√£o Serie A 2023**

Reading data 

```python

league = PROFESSIONAL_ARGENTINA_LEAGUE['league_name']

TOURNAMENT_2023 = f"./data/{league}-2023_matches.csv"
TOURNAMENT_2024 = f"./data/{league}-2024_matches.csv"

df_2023 = pd.read_csv(TOURNAMENT_2023)
df_2024 = pd.read_csv(TOURNAMENT_2024)
```

Filtering data less than 2.5 goals:

```python
less_goals_2023 = df_2023[df_2023['Total Goals'] < 2.5]
less_goals_2024 = df_2024[df_2024['Total Goals'] < 2.5]
```

Filtering data more than 2.5 goals:

```python
more_goals_2023 = df_2023[df_2023['Total Goals'] > 2.5]
more_goals_2024 = df_2024[df_2024['Total Goals'] > 2.5]
```

### **Statistics of matches with more/less than 2.5 goals** 

Percentage of matches with more/less than 2.5 goals:

```python
# Calculing the percentage of matches with less than 2.5 goals

less_goals_percentage_2023 = len(less_goals_2023) / len(df_2023) * 100
more_goals_percentage_2023 = 100 - less_goals_percentage_2023

less_goals_percentage_2024 = len(less_goals_2024) / len(df_2024) * 100
more_goals_percentage_2024 = 100 - less_goals_percentage_2024

labels = ['Less than 2.5 goles', 'More than 2.5 goles']
colors = ['#ff9999','#66b3ff']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.pie([less_goals_percentage_2023, more_goals_percentage_2023], 
        labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
ax1.set_title('Less/More than 2.5 goals in 2023')

ax2.pie([less_goals_percentage_2024, more_goals_percentage_2024], 
        labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
ax2.set_title('Less/More than 2.5 goals in 2024')

plt.show()
```

Teams with more local results with less than 2.5 goals:

```python
home_less_result_2023 = less_goals_2023['Home Team'].value_counts()
away_less_result_2023 = less_goals_2023['Away Team'].value_counts()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

home_less_result_2023.plot(kind='bar', ax=ax1)
ax1.set_title('Home Team with less than 2.5 goals in 2023')

away_less_result_2023.plot(kind='bar', ax=ax2)
ax2.set_title('Away Team with less than 2.5 goals in 2023')
```

```python
home_less_result_2024 = less_goals_2024['Home Team'].value_counts()
away_less_result_2024 = less_goals_2024['Away Team'].value_counts()

fg, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
home_less_result_2024.plot(kind='bar', ax=ax1, color='red')
ax1.set_title('Home Team with less than 2.5 goals in 2024')

away_less_result_2024.plot(kind='bar', ax=ax2, color='red')
ax2.set_title('Away Team with less than 2.5 goals in 2024')
```

## **Statistics of home/away teams goals**

Average goals of home/away teams 2023:

```python
home_goals_avg_2023 = df_2023.groupby('Home Team')['Home Score'].mean().sort_values(ascending=True)
away_goals_avg_2023 = df_2023.groupby('Away Team')['Away Score'].mean().sort_values(ascending=True)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

home_goals_avg_2023.plot(kind='bar', x='Home Team', y='Home Score', ax=ax1)
ax1.set_title('Home Team goals average in 2023')

away_goals_avg_2023.plot(kind='bar', x='Away Team', y='Away Score', ax=ax2)
ax2.set_title('Away Team goals average in 2023')
```

average goals of home/away teams 2024:

```python
home_goals_avg_2024 = df_2024.groupby('Home Team')['Home Score'].mean().sort_values(ascending=True)
away_goals_avg_2024 = df_2024.groupby('Away Team')['Away Score'].mean().sort_values(ascending=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

home_goals_avg_2024.plot(kind='bar', x='Home Team', y='Home Score', ax=ax1, color='red')
ax1.set_title('Home Team goals average in 2024')

away_goals_avg_2024.plot(kind='bar', x='Away Team', y='Away Score', ax=ax2, color='red')
ax2.set_title('Away Team goals average in 2024')
```
